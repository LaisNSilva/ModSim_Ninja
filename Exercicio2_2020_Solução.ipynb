**2. Gerando uma lista dinâmica em Python**

Na célula seguinte, implemente um código que preencha uma lista denominada “numerosPares”, de 100 posições (do índice 0 até o índice 99). Essa lista deve conter todos os primeiros cem números pares inteiros a partir do número zero. Porém, ela deve ser gerada usando-se a seguinte equação a diferenças:

$N(t+1)=N(t)+2$

Você deve usar uma estrutura de loop (como o “while”, por exemplo) para atualizar os valores da lista. Imprima a lista toda para se certificar que está correta. Imprima também o tamanho da lista (use a instrução “len(numerosPares)” para isso).


```python
#crie aqui seu código do item 2
i=1
numerosPares=100*[0]
while i<=99:
    numerosPares[i]=numerosPares[i-1]+2
    i=i+1
print(numerosPares)
print (len(numerosPares))
    
```

**3b.** Implementando as populações da Inglaterra, Escócia, País de Gales e Irlanda do Norte  em Python.

Na célula seguinte, implemente as equações a diferenças usadas no Exercício 1. Gere um gráfico de população em função do tempo para cada país (não esqueça de títulos, rótulo de eixos, cores etc).

**ATENÇÃO:** Os quatro estoques evoluem conjuntamente no tempo! Note, por exemplo, que a população da Inglaterra em (t+1) depende da população da Inglaterra em (t), mas também das populações da Escócia e do País de Gales em (t)! Como implementar, em um único loop, as quatro equações a diferenças?

```python
#crie aqui seu código do item 3b
import matplotlib.pyplot as plt
%matplotlib inline

#Definendo variáveis iniciais da população de cada país
I0 = 50E6
E0 = 5E6
G0 = 3E6
Ir0 = 2E6

#Inicializando listas de cada país
I = [0]*100
E = [0]*100
G = [0]*100
Ir = [0]*100

#Redefiniendo o primeiro elemento de cada lista
I[0]=I0
E[0]=E0
G[0]=G0
Ir[0]=Ir0

#Gerando a listas para cada país

i = 0
while i<99:
    I[i+1]=I[i]-0.01*I[i]+0.05*E[i]-2000
    E[i+1]=E[i]-0.05*E[i]+0.01*I[i]+1000
    G[i+1]=G[i]
    Ir[i+1]=Ir[i]+1000
    i=i+1
    
#Criando a lista Tempo
Tempo = list(range(0,100))

#Criando os gráficos Pop vs Tempo

plt.plot(Tempo,I,'ro')
plt.title('População da Inglaterra')
plt.xlabel("Anos")
plt.ylabel("População")
plt.grid(True)
plt.show()   

plt.plot(Tempo,E,'go')
plt.title('População da Escócia')
plt.xlabel("Anos")
plt.ylabel("População")
plt.grid(True)
plt.show()   

plt.plot(Tempo,G,'bo')
plt.title('População da País de Gales')
plt.xlabel("Anos")
plt.ylabel("População")
plt.grid(True)
plt.show()   

plt.plot(Tempo,Ir,'co')
plt.title('População da Irlanda do Norte')
plt.xlabel("Anos")
plt.ylabel("População")
plt.grid(True)
plt.show()   



```

**3c**. Gere o gráfico de ΔP em função de P para ao menos um dos países do Reino Unido. 

```python
#crie aqui seu código do item 3c

#Criando as listas Delta

DeltaI = [0]*99
DeltaE = [0]*99
DeltaG = [0]*99
DeltaIr = [0]*99


i=0
while i<99:
    DeltaI[i]=I[i+1]-I[i]
    DeltaE[i]=E[i+1]-E[i]
    DeltaG[i]=G[i+1]-G[i]
    DeltaIr[i]=Ir[i+1]-Ir[i]
    i=i+1

#Eliminando o ultimo elemento das listas de I, E, G, Ir
    
del I[99] 
del E[99]
del G[99]
del Ir[99]

#Criando os gráficos DeltaP vs P

plt.plot(I,DeltaI,'ro')
plt.title('Inglaterra: DeltaP vs P')
plt.xlabel("População")
plt.ylabel("Delta População")
plt.grid(True)
plt.show()   

plt.plot(E,DeltaE,'go')
plt.title('Escócia: DeltaP vs P')
plt.xlabel("População")
plt.ylabel("Delta População")
plt.grid(True)
plt.show()   

plt.plot(G,DeltaG,'bo')
plt.title('País de Gales: DeltaP vs P')
plt.xlabel("População")
plt.ylabel("Delta População")
plt.grid(True)
plt.show()   

plt.plot(Ir,DeltaIr,'co')
plt.title('Irlanda do Norte: DeltaP vs P')
plt.xlabel("População")
plt.ylabel("Delta População")
plt.grid(True)
plt.show()   
```
