**Exercício 3 - Derivadas, Equações Diferenciais e o Método de Euler**
   
Neste exercício, vamos entender melhor os conceitos de derivada e de equações diferenciais. Primeiramente, você já deve estar se acostumando com o conceito de derivada, que chamamos de "taxa de variação instantânea". Você já deve ter entendido que a taxa de variação temporal (média) de uma função no intervalo de tempo $t_1$ e $t_2$ é dada por:
   
$\frac{\Delta f}{\Delta t}=\frac{f(t_2)-f(t_1)}{t_2-t_1}$
   
Pois bem, já está ficando bem chato repetir que você ainda vai aprender em MatVar que quando $t_2$ é muito próximo de $t_1$ a taxa de variação será chamada de derivada (já sacou, né?). Mais uma vez:
   
$\frac{df}{dt}=\lim_{t_2 \to t_1}\frac{f(t_2)-f(t_1)}{t_2-t_1}$
   
Você deve estar muito ansioso(a) para aprender como calcular algebricamente a derivada de vários tipos de funcões, mas, por enquanto, vamos trabalhar o conceito da taxa de variação em um ponto, e construir uma série temporal de taxas para vários pontos. Ou seja, se começarmos a calcular as taxas de variação de uma determinada função quanto $t=0$ e formos incrementando esse $t$ a passos bem pequenos, poderemos calcular a taxa de variação a cada intervalo $\Delta t$, construindo uma lista de taxas de variação. A essa lista de taxas de variação, daremos o nome de "função derivada".


**Item 1 - Implementando numericamente a função derivada**

Vamos começar implementando uma função bem simples e muito conhecida por você: $f(t)=t^2$

Na célula abaixo, escreva uma função em Python para a implementação dessa função matemática. Perceba que estamos pedido apenas para **implementar** a função, não para executá-la!

```python
import math
#implementação da função
def f(t):
    #implemente aqui sua função
    y=t**2
    return y
```

1a) Para se certificar de que sua função funciona, plote um gráfico de $f(t)$ para $t$ variando de 0 a 10 com intervalos de tempo de 0,01 $s$. Use a função **arange** do módulo **numpy** para criar uma lista de tempo. Note que, na célula abaixo, você não precisa implementar a função novamente (ela foi implementada na célula anterior). Na célula abaixo, você vai **somente executá-la**.

```python
#implementação do gráfico da função f(t)
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

y=[]
delta_t=0.01
T=np.arange(0,10,delta_t)

for t in T:
    y.append(f(t))


plt.plot(T, y)
plt.axis([0, max(T), 0, max(y)])
plt.ylabel('f(t)')
plt.xlabel('t')
plt.title(r'Função f(t)')
plt.grid(True)
plt.show()
```

1b) Agora que sua função já está funcionando, vamos implementar o cálculo das taxas de variação de $f(t)$ para um determinado intervalo de tempo. Para isso, vamos criar uma função chamada **TaxadeVariacao(t, delta_t)**, que deve retornar a taxa de variação no instante t considerando um determinado $\Delta t$. Para isso, use a função f(t) que você já implementou!

```python
def TaxadeVariacao(t, delta_t):
    #implemente aqui sua função
    taxa = (f(t+delta_t)-f(t))/delta_t
    return taxa
```

1c) Agora vamos construir o gráfico da série temporal de f(t) e o da taxa de variação $\frac{\Delta f}{\Delta t}(t)$. Para isso, vamos criar duas listas vazias, uma denominada **y**, com os valores de f(t), e outra chamada **tx_var**, com os valores de $\frac{\Delta f}{\Delta t}(t)$. Use inicialmente o mesmo $\Delta t$ de 0,01 s com t entre 0 e 10.

```python
import matplotlib.pyplot as plt
import math
%matplotlib inline

y=[]
tx_var=[]
delta_t=0.01
T=np.arange(0,10,delta_t)

for t in T:
    #função
    y.append(f(t))
    #derivada da função
    tx_var.append(TaxadeVariacao(t, delta_t))
    
plt.plot(T, y, label='f(t)')
plt.plot(T, tx_var, label='df/dt')
plt.axis([0, max(T), min(y), 30])
plt.ylabel('f(t)')
plt.xlabel('t')
plt.title(r'Função f(t)')
plt.grid(True)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()   
    
```

```python
import math
#implementação da função
def f2(t):
    #implemente aqui sua função
    y=math.exp(t)
    return y
```

```python
import math
#implementação da função
def f3(t):
    #implemente aqui sua função
    y=math.cos(t)
    return y
```

```python
def TaxadeVariacaoGenerica(funcao, t, delta_t):
    #implemente aqui sua função
    t2=[t+delta_t]
    t1=[t]
    taxa = (funcao(*t2)-funcao(*t1))/delta_t
    return taxa
```

```python
import matplotlib.pyplot as plt
import math
%matplotlib inline


y=[]
y2=[]
y3=[]
tx_var=[]
tx_var2=[]
tx_var3=[]
delta_t=0.01
T=np.arange(0,10, delta_t)

for t in T:
    #função
    y.append(f(t))
    tx_var.append(TaxadeVariacaoGenerica(f, t, delta_t))
    #função 2
    y2.append(f2(t))
    tx_var2.append(TaxadeVariacaoGenerica(f2, t, delta_t))
    #função 3
    y3.append(f3(t))
    tx_var3.append(TaxadeVariacaoGenerica(f3, t, delta_t))
    
plt.plot(T, y, label='f(t)')
plt.plot(T, tx_var, label='df/dt')
plt.axis([0, max(T), min(y), max(y)])
plt.ylabel('f(t)')
plt.xlabel('t')
plt.title(r'Função f(t)')
plt.grid(True)
plt.legend()
plt.show()

plt.plot(T, y2, label='f2(t)')
plt.plot(T, tx_var2, label='df2/dt')
plt.axis([0, max(T), min(y2), max(y2)])
plt.ylabel('f(t)')
plt.xlabel('t')
plt.title(r'Função f(t)')
plt.grid(True)
plt.legend()
plt.show()

plt.plot(T, y3, label='f3(t)')
plt.plot(T, tx_var3, label='df3/dt')
plt.axis([0, max(T), min(y3), max(y3)])
plt.ylabel('f(t)')
plt.xlabel('t')
plt.title(r'Função f(t)')
plt.grid(True)
plt.legend()
plt.show()    
```

**Item 2 - Resolvendo mais uma equação diferencial pelo método de Euler, mas com uma taxa de variação que também é função do tempo!**

Vamos imaginar que queiramos saber qual a solução de uma equação diferencial do tipo $\frac{dy}{dt}=f(y, t)$.

Por exemplo, uma equação que seja dada pela expressão abaixo:

$\frac{dy}{dt}=a\cdot y\cdot[1+2\cdot sin(\omega \cdot t)]$

Essa equação representa a evolução diária da população de bactérias em um determinado meio, que depende da quantidade de bactéricas existentes e de uma função senoidal que representa a sazonalidade do processo.

Pois bem, vamos encontrar qual função $y(t)$ é solução da equação diferencial acima. Para isso, vamos usar, mais uma vez, o método de Euler.

Primeiramente, implemente abaixo uma função chamada **EquacaoDiferencial(y, t, a, w)**, que recebe os quatro parâmetros indicados, nessa ordem. Essa função deve retornar o lado direito da equação $dy/dt$.

```python
import math
def EquacaoDiferencial(y, t, a, w):
    #implemente sua função aqui
    dydt=a*y*(1+2*math.sin(w*t))    
    return dydt
```

Agora que sua função **EquacaoDiferencial** está implementada, vamos calcular a série temporal da função y(t) que satisfaz a equação diferencial que estamos tentando resolver. Os valores da função y(t) estarão numa lista, com cada elemento correspondendo a cada instante de tempo considerado. Abaixo, implemente o código do método de Euler executando a função **EquacaoDiferencial** e usando os seguintes parâmetros:

$y(0) = 5$ (Esta é a população inicial de bactérias)

$a = 0,007$

$\omega=0,05$

$\Delta t=0,001$ dias (nosso $\Delta t$ é um milésimo de um 1 dia)

Para sua implementação, use 400 dias de simulação.

Qual a população de bactérias após 1 ano (365 dias)?


```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
y=[5]
w=0.05
a=0.007
delta_t=0.001
T=np.arange(0,400, delta_t)

for i in range(0,len(T)-1):
    #calculamos a taxa de variação
    taxa = EquacaoDiferencial(y[i], T[i], a, w)
    #equação a diferenças
    y.append(y[i]+taxa*delta_t)

plt.plot(T, y)
plt.axis([0, max(T),0, 100])
plt.ylabel('f(t)')
plt.xlabel('t')
plt.title(r'Solução da Equação Diferencial f(t)')
plt.grid(True)
plt.show()   

#iteracao no dia 365
iteracao_365 = int(365/delta_t)

print('A população na iteração 365 é:', y[iteracao_365])

```
