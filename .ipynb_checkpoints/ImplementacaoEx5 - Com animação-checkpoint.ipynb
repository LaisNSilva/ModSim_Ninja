```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
from math import *

%matplotlib inline

m=0.2
L0=0.3
k=10
g=10
R = 0.1
A = pi * R**2
rho = 1
Cd = 1.5

def EqDif(S, t, resistAr):
    X= S[0]
    Vx=S[1]
    Y= S[2]
    Vy=S[3]    
    V = sqrt(Vx**2 + Vy**2)

    if V > 0.01 and resistAr:
        Far = 1/2 * rho * (Vx**2 + Vy**2) * Cd * A
        Farx = Far * Vx/V
        Fary = Far * Vy/V
    else:
        Farx = 0
        Fary = 0
        
    dXdt=Vx
    dVxdt=-k*(sqrt(X**2+(-Y)**2)-L0)*(X/sqrt(X**2+(-Y)**2))*(1/m)-Farx*(1/m)
    dYdt=Vy
    dVydt=k*(sqrt(X**2+(-Y)**2)-L0)*(-Y/sqrt(X**2+(-Y)**2))*(1/m)-Fary*(1/m)-g    

    return [dXdt, dVxdt, dYdt, dVydt]
    
S0=[0.55,0,0,0]
tempo=np.arange(0, 20,  1e-2)
# com resistência do ar
Sol=odeint(EqDif, S0, tempo, args=(True,))
plt.plot(Sol[:,0], Sol[:,2], 'red', label='Com resistência do ar')
plt.plot(0, 0, 'ro')
plt.axis('equal')
# sem resistência do ar
Sol2=odeint(EqDif, S0, tempo, args=(False,))
plt.plot(Sol2[:,0], Sol2[:,2], 'blue', label='Sem resistência do ar')
plt.plot(0, 0, 'bo')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()

plt.plot(tempo, Sol[:,2], 'r', label='Com resistência do ar')
plt.plot(tempo, Sol2[:,2], 'b', label='Sem resistência do ar')
plt.legend()
plt.show()

plt.plot(tempo, Sol[:,0], 'r', label='Com resistência do ar')
plt.plot(tempo, Sol2[:,0], 'b', label='Sem resistência do ar')
plt.legend()
plt.show()
    
    
```

```python
get_ipython().magic('matplotlib qt5')
import matplotlib.patches as patches
from matplotlib import animation


#este código faz com que o matplolib abra uma janela popup
%matplotlib qt5

#obtém solução do ODEINT
x=Sol[:,0]
y=Sol[:,2]

#determina o tamanho da janela e iguala as escalas x e y
fig = plt.figure(figsize=(20,10), facecolor='gray')
plt.axis('equal')
ax = fig.add_subplot(111)
#dimensão dos eixos
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 0.2)
ax.grid(True)

#especificando as posições iniciais, tamanho e cor da bolinha animada
Bola = patches.Circle((0, 0), 0.05, fc='blue')
VetorP, = ax.plot([],[], 'b-', lw=1)
Base, = ax.plot([-1, 1],[0,0], 'black', lw=4)

def init():
    #inclui as figuras que serão desenhadas    
    ax.add_patch(Bola)   
    VetorP, = ax.plot([],[], 'r-', lw=1)
    return None

def animate(i):
    #animação da bolinha    
    Bola.center=(x[i], y[i])  
    VetorP.set_data([0, x[i]], [0, y[i]])
    return None

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=len(y), interval=10, blit=False)

plt.show()
```
